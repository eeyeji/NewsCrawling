```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import time
from collections import Counter
import pickle
from tqdm import tqdm

import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc  
font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
#(빈)크롬창 열기
driver = webdriver.Chrome(ChromeDriverManager().install())

#url_list만들기
url_list = []

#sbs뉴스: 최신기사 페이지 별 창 열기
for i in range(1, 35):
    
    page_url = f'https://news.sbs.co.kr/news/search/main.do?query=%EC%A1%B8%EC%9D%8C%EC%9A%B4%EC%A0%84&pageIdx={i}&searchOption=1&collection='
    #page_url = f'https://news.sbs.co.kr/news/newsflash.do?plink=NEW&cooper=SBSNEWSMAIN&pageIdx={i}'
    driver.get(page_url)
    #driver.maximize_window()
    time.sleep(1)
    
    #기사 리스트(행) 별 url 저장-href 
    soup = bs(driver.page_source, 'html.parser')
    #url = soup.select('a.news')
    url = soup.select('a.psil_link')
    for j in url:
        #url_list.append('https://news.sbs.co.kr/'+j['href'])
        url_list.append(j['href'])
len(url_list)
```

```python
title_list=[]
contents_list=[]

for i in url_list:
    driver.get(i)
    time.sleep(1)
    soup = bs(driver.page_source, 'html.parser')
    #제목
    title = soup.select('h2.blind.ir')
#    for i in title:
#         title_list.append(i.text)
    title_list.append(title[0].text)
    
    #내용
    contents = soup.select('div.text_area')
#     for i in content:
#         contents_list.append(i.text)

#리스트에 쌓인 내용만 뽑기위해서 인덱스 사용, 즉 []표시 지우기 위해! 
    contents_list.append(contents[0].text)
```

```python
#필요없는 표시들 지우기
result_list =[]
for i in contents_list:
    result = i.replace("\n ", "")
    result = result.replace("\n\n", "")
    result = result.replace(" \n", "")
    result = result.replace("'\n'", "")
    result = result.replace("\xa0", "")
    result = result.replace("'\'", "")
    result = result.replace("사진", "")
    result_list.append(result)

result_list
```

```python
data = {'기사제목' : title_list, '기사내용' : result_list}
sbsNews = pd.DataFrame(data)
sbsNews.head()
```

```python
#제목과 내용하나로 묶기
sbsNews['sents'] = sbsNews['기사제목'] + ' ' + sbsNews['기사내용']
sbsNews.head()
```

```python
sbsNews.rename(columns={'기사제목':'title', '기사내용':'contents'}, inplace=True)
```
